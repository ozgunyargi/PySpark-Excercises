```python
# Filter out the data that only date and rate information will be hold. Remove the dates that does not contain rate information

def get_date_and_rate(myline):

  splitted_line = myline.split("\t")
  return (splitted_line[1], splitted_line[2])

rdd_dollar_ = rdd_dollar.map(lambda x: get_date_and_rate(x))
rdd_dollar_ = rdd_dollar_.filter(lambda x: x[1] != "") # Filter out unkonwn rate dates.
```

```python
# Re-index them

rdd_dollar_with_index = rdd_dollar_.zipWithIndex()
```

```python
# Re-add dropped 6 zeros after 31-12-2004

coversion_change_index = rdd_dollar_with_index.filter(lambda x: x[0][1] == "1,34").collect()[0][1]

def conversion(indx, mytuple):

  if mytuple[1] >= indx:
    return (mytuple[1], (mytuple[0][0], float(mytuple[0][1].replace(",", "."))*10**6))
  else:
    return (mytuple[1], (mytuple[0][0], float(mytuple[0][1].replace(".", "").replace(",", "."))))

rdd_dollar_with_index_  =  rdd_dollar_with_index.map(lambda x: conversion(coversion_change_index, x))
```

```python
# Copy the same dataset and drop the first index from the one. Drop the last index from the other

max_index = rdd_dollar_with_index_.collect()[-1][0]
rdd_dollar_first = rdd_dollar_with_index_.filter(lambda x : x[0]!= max_index)
rdd_dollar_second = rdd_dollar_with_index_.filter(lambda x : x[0]!= 0)
```

```python
# Organize the indexes that will be same with each other

def match_index(mytuple):
  return (mytuple[0]-1, mytuple[1])

rdd_dollar_second = rdd_dollar_second.map(lambda x: match_index(x))
```

```python
# Merge two rdds' together wrt index

rdd_merged = rdd_dollar_first.join(rdd_dollar_second)
rdd_sorted = rdd_merged.sortBy(lambda x: x[0])
```

```python
rdd_sorted.collect()[0]
```

```python
# Find the increases

def find_increase(mytuple):

  increase = (mytuple[1][1][1]-mytuple[1][0][1])/mytuple[1][0][1]

  return (mytuple[1][0][0], increase)

rdd_increases = rdd_sorted.map(lambda x: find_increase(x)).sortBy(lambda x: x[1], ascending=False)
```

```python
print("Top 5 greatest daily increase:\n"+"-"*len("Top 5 greatest daily increase:"))

for indx, i in enumerate(rdd_increases.collect()[:5]):
  print(indx+1, " => Day: ", i[0], " with percentage ", format(i[1]*100, ".3f"),"%", sep="")
```
